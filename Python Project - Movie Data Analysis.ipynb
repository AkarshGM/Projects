# Movie Data Analysis

```python
import pandas as pd
import ast
import json
```

```python

data = pd.read_csv('DS1_C8_V3_ND_Sprint2_Data Analysis Using Python_Dataset.csv')

df = pd.DataFrame(data)
df
```

```python
for row in range(len(df['genres'])):
    a = json.loads(df['genres'][row])
    b = [entry['name'] for entry in a]
    df.at[row,'genres'] = b
```

```python
df
```

```python
for row in range(len(df['production_companies'])):
    c = json.loads(df['production_companies'][row])
    d = [entry['name'] for entry in c]
    df.at[row,'production_companies'] = d
```

## Sort the data in descending order based on revenue and filter the top 500 movies. Find the measures of central tendency for the following columns using the filtered data: budget, revenue, runtime. Perform outlier analysis for the above three columns using box plot

```python
# Sort data in descending order based on revenue
sorted_df = df.sort_values(by='revenue', ascending=False)

# Filter the top 500 movies
filtered_df = sorted_df.head(500)

# Calculate measures of central tendency for budget, revenue, and runtime columns
budget_mean = filtered_df['budget'].mean()
budget_median = filtered_df['budget'].median()
budget_mode = filtered_df['budget'].mode().iloc[0]

revenue_mean = filtered_df['revenue'].mean()
revenue_median = filtered_df['revenue'].median()
revenue_mode = filtered_df['revenue'].mode().iloc[0]

runtime_mean = filtered_df['runtime'].mean()
runtime_median = filtered_df['runtime'].median()
runtime_mode = filtered_df['runtime'].mode().iloc[0]

# Display the measures of central tendency
print("Measures of Central Tendency for Budget:")
print("Mean:", budget_mean)
print("Median:", budget_median)
print("Mode:", budget_mode)

print("\nMeasures of Central Tendency for Revenue:")
print("Mean:", revenue_mean)
print("Median:", revenue_median)
print("Mode:", revenue_mode)

print("\nMeasures of Central Tendency for Runtime:")
print("Mean:", runtime_mean)
print("Median:", runtime_median)
print("Mode:", runtime_mode)
```

```python
# Perform outlier analysis using box plots
plt.figure(figsize=(10, 6))

# Box plot for budget
plt.subplot(1, 3, 1)
plt.boxplot(filtered_df['budget'], vert=False)
plt.title("Outlier Analysis - budget")

# Box plot for revenue
plt.subplot(1, 3, 2)
plt.boxplot(filtered_df['revenue'], vert=False)
plt.title("Outlier Analysis - revenue")

# Box plot for runtime
plt.subplot(1, 3, 3)
plt.boxplot(filtered_df['runtime'], vert=False)
plt.title("Outlier Analysis - runtime")

plt.tight_layout()
plt.show()
```
