# Analysis of Income Data

```python
import matplotlib.pyplot as plt          
from tabulate import tabulate             
import matplotlib as mat                   
import seaborn as sns                    
import pandas as pd                        
import numpy as np                     
import warnings
warnings.filterwarnings('ignore')
```

```python

data = pd.read_excel('DS1_C5_S2_Income_Data_Challenge.xlsx')

df = pd.DataFrame(data)
df
```

# Task 1

```python

def seprate_data_types(df):
    categorical = []
    continuous = []
    for column in df.columns:               
        if df[column].nunique() < 73:       
                                                    
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous 


categorical, continuous = seprate_data_types(df)        


from tabulate import tabulate
table = [categorical, continuous]
print(tabulate({"Categorical":categorical,
                "continuous": continuous}, headers = ["categorical", "continuous"]))
```

# Task 2

## 1. Analysis of Education with Income Bracket

```python
fig, ax=plt.subplots(figsize=(10,7))
sns.countplot(y="Education", hue="Income_Bracket", data=df)
plt.legend(bbox_to_anchor=(1,1));
```

##### Most of the workforce having HS Graduation earn less than or equal to 50K.

## 2. Analysis of Occupation with Income Bracket

```python
fig, ax=plt.subplots(figsize=(10,8))
sns.countplot(y="Occupation", hue="Income_Bracket", data=df)
plt.legend(bbox_to_anchor=(1,1));
```

##### Maximum type of Occupations come under less than 50K income Category.

## 3. Analysis of Gender with Income Bracket

```python
fig, ax=plt.subplots(figsize=(10,7))
sns.countplot(x="Gender", hue="Income_Bracket", data=df)
plt.legend(bbox_to_anchor=(1.05,1));
```

##### Gender with income more than 50K and less than 50K is maximum in Male category.

## 4. Analysis of Marital Status with Income Bracket

```python
fig, ax=plt.subplots(figsize=(12,7))
sns.countplot(x="Marital_Status", hue="Income_Bracket", data=df)
plt.legend(bbox_to_anchor=(1.05,1));
```

##### Highest earning Marital status category is 'Married' crossing 50K income bracket.

## 5. Analysis of Workclass with Income Bracket

```python
fig, ax=plt.subplots(figsize=(10,7))
sns.countplot(x="WorkClass", hue="Income_Bracket", data=df)
plt.legend(bbox_to_anchor=(1.05,1));
```

##### Maximum number of Private Workclass earn less than 50K as well as more than 50K compatred to all other workforces.
