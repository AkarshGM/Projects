## Automobile Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('DS1_C4_S5_Car_Data_Challenge.xlsx')

df = pd.DataFrame(data)
df
```

```python
# Task 1

a = df.sort_values('City_Mileage_km_litre', ascending = False)
top = a[['Make', 'City_Mileage_km_litre']]
top_10 = top.head(10)
top_10
```

```python
# plot bar graph

top_10.plot.barh('Make');
```

```python
top_10.boxplot(column = ['City_Mileage_km_litre'], by = ['Make']);
```

## There are no outliers in the above output. The mileage range of maruti suzuki is lesser than Mahindra.

```python
# Task 2 


b = df.sort_values('City_Mileage_km_litre', ascending = False)
nonelectric = b[(b['Fuel_Type'] != 'Electric')]
top = nonelectric[['Make', 'City_Mileage_km_litre']]
top_20 = top.head(20)
top_20
```

```python
# Top 10 non electric car manufacturers based on city mileage

top_20.plot.barh('Make');
```

```python
# Task 3

body_type = df.groupby(by = ['Body_Type'])['Body_Type'].count().nlargest(5)
body_type
```

```python
# plotting horizontal bar graph

body_type.plot.barh( title = 'Body Type Vs Count of cars');
```

```python
# Task 4

cars = df[(df['Make'] == 'Hyundai') | (df['Make'] == 'Mahindra') | (df['Make'] == 'Renault') | (df['Make'] == 'Skoda') ]
mileage = cars[['Make', 'City_Mileage_km_litre', 'Highway_Mileage_km_litre']]
mileage
```

```python
# Calculating Mean of city mileage of Hyundai, Mahindra, Renault, Skoda cars

city_mileage_mean = pd.pivot_table(mileage, index = ['Make'], values = ['City_Mileage_km_litre'], aggfunc = 'mean')
city_mileage_mean
```

```python
# Calculating Median of city mileage of Hyundai, Mahindra, Renault, Skoda cars

city_mileage_median = pd.pivot_table(mileage, index = ['Make'], values = ['City_Mileage_km_litre'], aggfunc = 'median')
city_mileage_median
```

```python
# Calculating Mean of Highway mileage of Hyundai, Mahindra, Renault, Skoda cars

highway_mean = pd.pivot_table(mileage, index = ['Make'], values = ['Highway_Mileage_km_litre'], aggfunc = 'mean')
highway_mean
```

```python
# Calculating Median of Highway mileage of Hyundai, Mahindra, Renault, Skoda cars

highway_median = pd.pivot_table(mileage, index = ['Make'], values = ['Highway_Mileage_km_litre'], aggfunc = 'median')
highway_median
```

```python
# Task 5

mean_median = pd.concat([city_mileage_mean['City_Mileage_km_litre'], city_mileage_median['City_Mileage_km_litre']], keys= ['City Mileage Mean', 'City Mileage Median'], axis=1)
mean_median
```

```python
# Plot stacked column chart

mean_median.plot(kind = 'bar',
                stacked = True,
                title = 'Mean & Median Values of City Mileage',
                xlabel = 'Car Manufacturers',
                ylabel = 'City Mileage in km');
```

```python
# Task 6 - finding null values in mileage

df[['City_Mileage_km_litre', 'Highway_Mileage_km_litre']].isnull().sum()
```

```python
# Imputing missing values with mean in city mileage column

df['City_Mileage_km_litre'] = df['City_Mileage_km_litre'].fillna(df['City_Mileage_km_litre'].mean())
df['City_Mileage_km_litre'].isnull().sum()
```

```python
# Imputing missing values with mean in Highway mileage column

df['Highway_Mileage_km_litre'] = df['Highway_Mileage_km_litre'].fillna(df['Highway_Mileage_km_litre'].mean())
df['Highway_Mileage_km_litre'].isnull().sum()
```

```python
df[['City_Mileage_km_litre', 'Highway_Mileage_km_litre']].isnull().sum()
```

```python
average_mileage = pd.pivot_table(df, index = 'Make', values = ['City_Mileage_km_litre', 'Highway_Mileage_km_litre'], aggfunc = 'mean').sort_values(by=['City_Mileage_km_litre'],ascending=True)
average_mileage
```

```python
# Plotting horizontal bar chart for average of city and highway mileage sorted in ascending order of  city mileage

average_mileage.plot(kind = 'barh',
                title = 'Average City and Highway Mileage of car makers',
                xlabel = 'Car Manufacturers',
                figsize=(14,25));
```

```python
# Task 7

df1 = df[df['Displacement'].notna()]
c = df1.sort_values('City_Mileage_km_litre', ascending = False)
topp = c[['Displacement', 'City_Mileage_km_litre']]
displacement = topp.head(10)
displacement
```

```python
# Plot line chart

displacement.plot(kind = 'line',
                 title = 'Displacement Vs City Mileage',
                 xlabel = 'Mileage',
                 ylabel = 'Displacement');
```

```python
displacement.plot(y= 'Displacement', x = 'City_Mileage_km_litre', kind = 'line');
```

## Interpretation: Displacement of a car is negatively co-related to Mileage. Higher the Displacement, Lesser is the Mileage and Vice versa
