## Employee Data Analysis

```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('DS1_C4_S5_Employee_Data_Practice.xlsx')
df = pd.DataFrame(data)
df
```

```python
# Task 1

xy = df.groupby('Department').sum()[['Annual Salary ($)']]
xy

            
```

```python
# Ploting the graph


xy.plot.bar(title = 'Department-wise Annual Salary', ylabel = 'Salary', color = 'g', width = 0.4);
```

```python
# Task 2

dept = df.groupby(by = ['Department', 'Gender'])['Gender'].count()
dept
```

```python
# Creating dataframe
dataframe = pd.DataFrame({'Department': ['Fincance', 'HR', 'IT', 'Sales'], 'Male Count': [11,4,12,16], 'Female Count': [2,11,10,4]})
dataframe
```

```python
dataframe.plot(x= 'Department',
              kind = 'bar',
              stacked = False,
              title = 'Gender-wise count of departments', xlabel = 'Department', ylabel = ' Employee count');
```

```python
# Task 3

IT = df[(df['Department'] == 'IT ') & (df['Work_Experience'] == 0)]
print("The Average salary of a Fresher in IT department is:", IT['Annual Salary ($)'].mean())
IT
```

```python
# Task 4

sal = df.groupby('Department').sum()[['Annual Salary ($)']]
sal
```

```python
dep = ['Finance', 'HR', 'IT', 'Sales']
salary = [790000, 987000, 1282900, 1089000]

plt.pie(salary, labels = dep, autopct = '%.2f %%' )
plt.title('Dept-wise sum of Annual salaries')
plt.show()
```

```python
# Task 5

plt.title('Distribution of Employee count by Age group')
plt.xlabel('Age group')
plt.ylabel('Employee Count')
plt.hist(df['Age'], bins = 5,edgecolor = 'r');
```

```python
# Task 6

trend = df[['Annual Salary ($)', 'Work_Experience']]
trend = trend.sort_values(by = 'Work_Experience')
trend
```

```python
# Scatter Plot
plt.xlabel('Work Experience')
plt.ylabel('Annual Salary')
plt.title('Salary Trend based on Employee experience')
plt.scatter(trend['Work_Experience'], trend['Annual Salary ($)']);
```

```python
# Task 7

df.boxplot(column = ['Annual Salary ($)'], by = 'Department')
plt.tight_layout()
```
