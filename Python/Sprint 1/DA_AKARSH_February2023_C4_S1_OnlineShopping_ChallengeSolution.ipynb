## Online shopping Data Analysis

```python
import numpy as np
from numpy import genfromtxt

Data = genfromtxt('DS1_C4_S1_Shopping_Data_Challenge.csv', delimiter = ',', skip_header = 1)
print(Data)
```

```python
# Task 1

#Assigning Variables to each column

cust_id = Data[:,0]
age = Data[:,1]
annual_income = Data[:,2]
spend_score = Data[:,3]


index = np.where(spend_score>80)
score_morethan_80 = spend_score[index]
print("---------------------------------------------------------------")
print("*Details of the customers whose spending score is more than 80*")
print("---------------------------------------------------------------")
print("\nCustomer ID:\n",cust_id[index])
print("\nAge:\n",age[index])
print("\nAnnual Income:\n",annual_income[index])
print("\nSpending Score:\n",score_morethan_80)

# Display total count of these customers

print("=========================================================================")
print("\nTotal count of customers whose spending score is more than 80 is:", cust_id[index].size)
```

```python
# Task 2

indx = np.where((age>20) & (age<24))
age_group = age[indx]

print("--------------------------------------------------------")
print("*Details of the customers whose age is between 20 to 24*")
print("--------------------------------------------------------")
print("\nCustomer ID:\n",cust_id[indx])
print("\nAge:\n",age_group)
print("\nAnnual Income:\n",annual_income[indx])
print("\nSpending Score:\n", spend_score[indx])
```

```python
# Task 3

lowest_index = np.argmin(annual_income)
highest_index = np.argmax(annual_income)

print("Customer IDs of lowest and highest annual income is:\n", cust_id[lowest_index], "and", cust_id[highest_index], "respectively.")
```

```python
# Task 4 

indxx = np.where((age<45) & (annual_income>35) & (spend_score>75))

print("---------------------------------------------------------------------------------------------------------------")
print("Details of the customers whose spending score is more than 75, age less than 45, annual income greater than 35K")
print("---------------------------------------------------------------------------------------------------------------")
print("\nCustomer ID:\n",cust_id[indxx])
print("\nAge:\n",age[indxx])
print("\nAnnual Income:\n",annual_income[indxx])
print("\nSpending Score:\n",spend_score[indxx])
```

```python
# task 5

# Generating random inputs from $10 to $50 for 200 employees 
gift_coupons = np.random.randint(10,51, 200)
print("Random Gift Coupons from the company:\n\n", gift_coupons)
# Adding new column to main data

new_data = np.column_stack((Data, gift_coupons))
print("\nData after adding new 5th column as 'Gift Coupons' is:\n", new_data)
```
