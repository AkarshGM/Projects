## University Admission Decision

```python
import numpy as np
```

```python
# Task 1

from numpy import genfromtxt

adm_data = genfromtxt('DS1_C4_S1_AdmissionPredict_Data_Practice.csv', delimiter = ',', skip_header = 1)
print(adm_data)
```

```python
# Task 2

GRE = adm_data[:,1]
print("GRE Score of all students:\n\n", GRE)
```

```python
# Task 3

index = np.where(GRE>320)
GRE_morethan_320 = GRE[index]
print("Students with GRE score greater than 320 is:\n\n", GRE_morethan_320)

# Find total number of students with GRE score more than 320

total = GRE_morethan_320.size
print("\n\nThe Total number of Students with GRE score greater than 320 is:", total)
```

```python
# Task 4 

TOEFL = adm_data[:,2]
print("TOEFL Score of all students:\n\n", TOEFL)

indx = np.where(TOEFL>110)
Tofl_morethan_110 = TOEFL[indx]
print("\n\nStudents with TOEFL score greater than 110 is:\n\n", Tofl_morethan_110)

# Calculate chance of admission

chance = adm_data[indx, 8]
print("\n\nChance of Admission is:", chance)
avg = chance.mean()

percent = avg * 100
print("\n\nChance of admission who scores more than 110 in TOEFL is:", percent, "Percentage")
```

```python
# Task 5

research = adm_data[:, -2]
print("Research Experience:\n", research)

Indexx = np.where((GRE > 315) & (TOEFL >115) & (research == 1))
print("\n\nIndexes satisfying the condition:\n", Indexx)

#Finding serial number

serial = adm_data[:, 0]
print("\n\nSerial numbers:\n", serial)
serial_of_students = serial[Indexx]
print("\n\nIndexed Serial numbers are:\n", serial_of_students)
print("\n\nThe total number of students eligible for scholarship is:", serial_of_students.size)
```
