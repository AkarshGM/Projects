## Used Cars: Sales Data Analysis

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('DS1_C4_S4_Used_Car_Prices_Data_Challenge.csv')

df = pd.DataFrame(data)
df
```

```python
# There are (â€”) type of characters in csv file in color and interior columns
# After importing the csv file, it shows those cells as (—) in jupyter
# Hence, replacing those (—) cells with NaN so that those rows can be removed in next step using dropna

df=df.replace('—', np.nan)
df
```

```python
# Task 1 - checking NaN value counts

df.isnull().sum()
```

```python
df.dropna(how = 'any', inplace = True)      # dropping missing value rows 
```

```python
df.isnull().sum()     #Verifying if missing value rows are removed 
```

# Missing value rows are removed from the data above 

```python
# Task 2 - top 100 lowest priced cars from year 2014 and registered in CA

req = df[(df['year'] == 2014) & (df['state'] == 'ca')]
sorted_price = req.sort_values('mp')
top = sorted_price.head(100)
top
```

```python
# Task 3

cross = pd.crosstab(top.make, top.mp, rownames = ['Car Manufacturer'], colnames = ['Price'])
cross
```

```python
# Task 4

chevvy = top[top['make'] == 'Chevrolet']
chevvy
```

```python
# Applying 10% discount on original price of chevrolet cars

disc = chevvy.assign(Discounted_price = chevvy.mp * 0.9)
disc[['model', 'mp', 'Discounted_price']]
```

```python
# Task 5

interior =  df[df['year'] == 2014]
top_100 = interior.sort_values('mp')
b = top_100.head(100)
b
```

```python
print("Average price of all the 100 lowest priced cars from the year 2014 based on their interior colors:")
b.groupby(by = ['make', 'interior'])['mp'].mean()
```

```python
# Task 6

locate =  df.loc[(df['year'] >= 2010) & (df['year'] <= 2015)]
locate
```

```python
print("Color-wise availability of cars for each manufacturer from the year 2010-2015:")
locate.groupby(by = ['make', 'color'])['model'].count()
```

```python
# Task 7

bmw = locate[(locate['condition'] >= 4.5) & (locate['make'] == 'BMW')]
bmw
```

```python
print("Year-wise maximum expected price of BMW cars:")
bmw.groupby(by = ['year','make', 'model'])['mp'].max()
```

```python
# Task 8

suv = locate[(locate['body'] == 'SUV') & (locate['odometer'] <= 30000)]
sorted_suv = suv.sort_values('mp', ascending = False)
print("Top 100 SUVs with prices in descending order:")
sorted_suv.head(100)
```
