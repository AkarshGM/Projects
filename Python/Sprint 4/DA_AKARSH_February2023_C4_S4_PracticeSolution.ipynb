## Test Scores of Students

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('DS1_C4_S4_Test_Scores_Data_Practice.csv')

df = pd.DataFrame(data)
df
```

```python
# Task 1
# Finding null values 

df.isnull().sum()
```

```python
# Finding most frequently found data from the respective columns

df['gender'].value_counts()
```

```python
# Imputing data to gender column with Male

df['gender'] = df['gender'].fillna('Male')
print(df.isnull().sum())
```

```python
# Imputing data to test1 column with its mean

df['test1'] = df['test1'].fillna(df['test1'].mean())
print(df.isnull().sum())
```

```python
# Imputing data to test2 column with its median

df['test2'] = df['test2'].fillna(df['test2'].median())
print(df.isnull().sum())
```

```python
# Task 2 - finding duplicates and display them

df_dup = df[df.duplicated(keep='first')] 
df_dup
```

```python
# drop duplicates

df.drop_duplicates(inplace = True)
df
```

```python
# Task 3

df = df.replace(to_replace= ['Does not qualify', 'Qualifies for reduced/free '], value = ['No', 'Yes'])
df
```

```python
# Task 4

a = df.groupby(by = 'school').mean('test1')
a
```

```python
# Finding highest average test score

a['test1'].max()
```

```python
# displaying the name of the school with highest average test score

a.loc[a['test1']== 78.453125]
```

```python
# Task 5

pvt = pd.pivot_table(df, index = ['school'], values = ['test2'], aggfunc = 'max')
high = pvt.sort_values('test2', ascending = False)
high
```

```python
# Displaying top 5 schools with highest test 2 scores

high.head()
```
