```python
# Task 6 - Customers who are not married, are self-employed, live in non-urban area and have loan status as NO.

gender = df[(df['Married'] == 'No') & (df['Self_Employed'] == 'Yes') & (df['Property_Area'] != 'Urban') & (df['Loan_Status'] == 'N') ]
gender
```

```python
# Task 7

# For Urban Area
urban = df[df['Property_Area'] == 'Urban']
print("The Average Income of customers living in Urban Area is:", urban['ApplicantIncome'].mean())

# For Semiurban Area
semiurban = df[df['Property_Area'] == 'Semiurban']
print("\nThe Average Income of customers living in Semiurban Area is:", semiurban['ApplicantIncome'].mean())

# For Rural Area
rural = df[df['Property_Area'] == 'Rural']
print("\nThe Average Income of customers living in Rural Area is:", rural['ApplicantIncome'].mean())
```

```python
# Task 8 - Customers with property in rural and semiurban, applicant income more than 4000, co-applicant income
           # more than 1000 and loan status as NO.
loan = df[(df['Property_Area'] != 'Urban') & (df['ApplicantIncome'] > 4000) & (df['CoapplicantIncome'] >1000) & (df['Loan_Status'] == 'N')]
loan
```

```python
# Total count of such customers

print("Total Count of customers:", len(loan.index))
```

```python
# Updating the loan status of above mentioned customers to YES

loan.loc[loan['Loan_Status'] == 'N', 'Loan_Status'] = 'Y'
loan
```

```python
# Updating the above edited data to main data 

df.update(loan)

# Re-doing same task to check if the new data is updated and the count of such customers is now zero

df[(df['Property_Area'] != 'Urban') & (df['ApplicantIncome'] > 4000) & (df['CoapplicantIncome'] >1000) & (df['Loan_Status'] == 'N')]
```

```python
# The Average Income of above customers

print("The Average income of customers with 3+ dependants, living in Urban area with loan status YES is:",
      income['ApplicantIncome'].mean())

# Count of customers with 3+ dependants, living in Urban area with loan status YES

print("\nThe Count of customers with 3+ dependants, living in Urban area with loan status YES is:", len(income.index))
```
