## Restaurant Sales Data

```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
# import data
data = pd.read_csv('DS1_C4_S3_SalesData_Practice.csv')
data
```

```python
# Task 1

arrange = data.sort_values('Sales')
print("Sorted sales data in ascending order is:")
arrange
```

```python
# Task 2

sale = data[data.Sales >= 45000]
print("Sales data greater than or equal to 45000 is:")
sale
```

```python
# Task 3

# Data of California

cali = data[data.State == 'California']
print("Sales Data for California in ascending order is:")
cali.sort_values('Sales')
```

```python
# Data of Texas

TX = data[data.State == 'Texas']
print("Sales data for Texas in ascending order is:")
TX.sort_values('Sales')
```

```python
# Data of Montana

Montana = data[data.State == 'Montana']
print("Sales data for Montana in ascending order is:")
Montana.sort_values('Sales')
```

```python
# Task 4

TX_sales = TX['Sales'].sum()
cali_sales = cali['Sales'].sum()
Montana_sales = Montana['Sales'].sum()
print("Total sales in the state of Texas is:", TX_sales)
print("Total sales in the state of California is:", cali_sales)
print("Total sales in the state of Montana is:", Montana_sales)
```

```python
# Task 5

monthly_sales = cali.groupby('Month')['Sales'].sum()
highest_month = monthly_sales.idxmax()
lowest_month = monthly_sales.idxmin()

print('Month with the highest sales is:', highest_month)
print('Month with the lowest sales is:', lowest_month)
```

```python
# Task 6

data_2 = {'State': ['Alaska', 'Texas', 'California', 'Montana', 'New Mexico'], 'Sales': [174149, 214856, 184644, 186270, 161917]}
new_data = pd.DataFrame(data_2)
new_data
```

```python
# Task 7

total_sales = new_data.sort_values('Sales')

# Exporting as .csv file

total_sales.to_csv('total_sales.csv', index=False)
total_sales
```
