```python
import mysql.connector as sql
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
db = sql.connect(host = 'localhost', user = 'root', password = 'root123', database = 'hr')
```

```python
mycursor = db.cursor()
```

```python
# Task 1

countries_data = pd.read_sql_query('SELECT * FROM countries', db)
countries_data
```

```python
# Task 2

employees_data = pd.read_sql_query('SELECT * FROM employees', db)
employees_data
```

```python
locations_data = pd.read_sql_query('SELECT * FROM locations', db)
locations_data
```

```python
department_data = pd.read_sql_query('SELECT * FROM departments', db)
department_data
```

```python
df = pd.merge(employees_data, department_data, on = 'department_id')
df
```

```python
df1 = pd.merge(countries_data, locations_data, on = 'country_id')
df1
```

```python
df2 = pd.merge(df, df1, on = 'location_id')
df2
```

```python
# Country-wise employee count 

new_data = pd.pivot_table(df2, index = ['country_name'], values = ['employee_id'], aggfunc = 'count').reset_index('country_name')
new_data
```

```python
# Renaming columns

new_data.rename(columns={"country_name": "Country_name","employee_id": "Employee_count"}, inplace=True)
new_data
```

```python
# Pushing the above data to MySQL RDBMS

from sqlalchemy import create_engine

# credentials for database connection
hostname="localhost"
dbname="hr"      # Name of the database into which we want into insert the data 
uname="root"
pwd="root#123"

# Create SQLAlchemy engine to connect to MySQL Database
engine = create_engine("mysql+pymysql://{user}:{pw}@{host}/{db}"
.format(host=hostname, db=dbname, user=uname, pw=pwd))

# Convert dataframe to sql table                                   
new_data.to_sql('countrywise_count', engine, index=False) # countrywise_count is the name of the new table

```

```python
# Task 3 - city-wise employee count

city = pd.pivot_table(df2, index = ['city'], values = ['employee_id'], aggfunc = 'count').reset_index()
city
```

```python
# Plotting graph for the above data for city-wise employee count

city.plot.barh('city');
plt.title('City-wise Employee Count')
plt.legend(["Employee count"]);
```

```python
# Country-wise employee count

country = pd.pivot_table(df2, index = ['country_name'], values = ['employee_id'], aggfunc = 'count').reset_index()
country
```

```python
# Plotting graph for the above data for country-wise employee count

country.plot.barh('country_name');
plt.title('Country-wise Employee Count')
plt.legend(["Employee count"]);
```

```python
# Exporting csv file

city.to_csv('Task 3 city CSV output.csv')
country.to_csv('Task 3 country CSV output.csv')
```
